```python
def stylize_title(document):
    return add_border(center_title(document))


# Don't touch below this line


def center_title(document):
    width = 40  # fixed width for centered title
    title = document.split("\n")[0]  # assumes the title is the first line
    centered_title = title.center(width)
    return document.replace(title, centered_title)


def add_border(document):
    title = document.split("\n")[0]
    border = "*" * len(title)
    return document.replace(title, title + "\n" + border)


def test(document):
    print("Input:")
    print(document)
    print("----------------------------------------")
    print("Output:")
    print(stylize_title(document))
    print("========================================")


def main():
    test(
        """The Importance of FP
Learn how functional programming can change the way you think about code.
Benefits include immutability, simplicity, and composability."""
    )
    test(
        """DocToDoc: A Guide
Understanding the art of document conversion.
We write cool functional code to make it happen."""
    )


main()
```

```python
def add_doc(document, documents):
    prefix = f"{len(documents)}. "
    new_doc = prefix + document
    new_documents = documents + (new_doc,)
    return new_documents


# Don't touch below this line


def test(docs_to_add):
    documents = ()
    for doc in docs_to_add:
        print(f"> Adding: {doc}")
        documents = add_doc(doc, documents)
        for doc in documents:
            print(f"  - {doc}")
    print("=====================================")


def main():
    test(["hello there", "sonny", "how ya doing"])


main()
```

```python
def get_median_font_size(font_sizes):
    sorted_sizes = sorted(font_sizes)
    n = len(sorted_sizes)
    if n == 0:
        return None
    if n % 2 == 0:
        return (sorted_sizes[n // 2 - 1] + sorted_sizes[n // 2]) / 2
    else:
        return sorted_sizes[n // 2]

# Don't edit below this line


def test(font_sizes):
    median_font_size = get_median_font_size(font_sizes)
    print(
        "Median font size for font sizes {} is {}".format(font_sizes, median_font_size)
    )
    print("=====================================")


def main():
    test([10, 12, 14])
    test([9, 11, 16, 20])
    test([8, 8, 8])
    test([14, 18, 22, 30])
    test([6, 6, 6, 24, 24, 24])
    test([])


main()
```
