```python
def sum(nums):
    if len(nums) == 0:
        # we've reached the end of the list
        # stop the recursion and return 0
        return 0
    # add the first number to the sum of the rest of the numbers
    return nums[0] + sum(nums[1:])

print(sum([1,2,3,4,5]))
```

```python
def factorial_r(x):
    if x == 1:
        return x
    return x * factorial_r(x - 1)


# TESTS


run_cases = [
    (5, 120),
    (10, 3628800),
]

submit_cases = run_cases + [
    (1, 1),
    (2, 2),
    (3, 6),
    (6, 720),
    (8, 40320),
]


def test(input1, expected_output):
    print("---------------------------------")
    print(f"Inputs: {input1}")
    print(f"Expecting: {expected_output}")
    result = factorial_r(input1)
    print(f"Actual: {result}")
    if result == expected_output:
        print("Pass")
        return True
    print("Fail")
    return False


def main():
    passed = 0
    failed = 0
    for test_case in test_cases:
        correct = test(*test_case)
        if correct:
            passed += 1
        else:
            failed += 1
    if failed == 0:
        print("============= PASS ==============")
    else:
        print("============= FAIL ==============")
    print(f"{passed} passed, {failed} failed")


test_cases = submit_cases
if "__RUN__" in globals():
    test_cases = run_cases

main()
```

```python
def zipmap(keys, values):
    if len(keys) == 0 or len(values) == 0:
        return {}
    dict = zipmap(keys[1:], values[1:])
    dict[keys[0]] = values[0]
    return dict

run_cases = [
    (
        ["name", "age", "city"],
        ["Alice", 25, "Wonderland"],
        {"city": "Wonderland", "age": 25, "name": "Alice"},
    ),
    (
        ["name", "age", "city"],
        ["Bob", 30, "Atlantis"],
        {"city": "Atlantis", "age": 30, "name": "Bob"},
    ),
]

submit_cases = run_cases + [
    ([], [], {}),
    (
        ["key1", "key2", "key3"],
        ["value1", "value2", "value3"],
        {"key3": "value3", "key2": "value2", "key1": "value1"},
    ),
    (
        ["father", "sister", "brother"],
        ["Homer", "Lisa", "Bart"],
        {"brother": "Bart", "sister": "Lisa", "father": "Homer"},
    ),
    (
        ["country", "state"],
        ["USA", "California"],
        {"state": "California", "country": "USA"},
    ),
]


def test(keys, values, expected_output):
    print("---------------------------------")
    print(f"Inputs: {keys}, {values}")
    print(f"Expecting: {expected_output}")
    result = zipmap(keys, values)
    print(f"Actual: {result}")
    if result == expected_output:
        print("Pass")
        return True
    print("Fail")
    return False


def main():
    passed = 0
    failed = 0
    for test_case in test_cases:
        correct = test(*test_case)
        if correct:
            passed += 1
        else:
            failed += 1
    if failed == 0:
        print("============= PASS ==============")
    else:
        print("============= FAIL ==============")
    print(f"{passed} passed, {failed} failed")


test_cases = submit_cases
if "__RUN__" in globals():
    test_cases = run_cases

main()
```

```python
def count_nested_levels(nested_documents, target_document_id, level=1):
    for key in nested_documents:
        print(f"key={key},lvl={level}")
        if target_document_id == key:
            print(f"FOUND ON LEVEL: {level}")
            return level
        else:
            level += 1
            doc_level = count_nested_levels(nested_documents[key], target_document_id, level)
            print(f"dl={doc_level},lvl={level}")
            if doc_level == -1:
                level += doc_level
                continue
            return doc_level
    return -1

""" # COURSE SOLUTION
def count_nested_levels(nested_documents, target_document_id, level=1):
    for document_id in nested_documents:
        if document_id == target_document_id:
            return level
        found_level = count_nested_levels(
            nested_documents[document_id], target_document_id, level + 1
        )
        if found_level != -1:
            return found_level
    return -1
"""
```
